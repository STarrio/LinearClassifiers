```python
def char_converter(c):
    return{' ':0,'\n':0,'+':1,'#':1}[c]
```

```python
def digit_parser(path):
    file = open(path)
    digits=[]
    digit=[]
    idx=0
    for line in file:
        digit+=[char_converter(x) for x in line]
        idx+=1
        if(idx%28==0):
            digits.append(np.array(digit))
            digit=[]
    return digits
```
